```python
import numpy as np
import matplotlib.pyplot as plt

def tukey(r, c):
    if np.abs(r) <= c:
        return (c**2 / 6) * (1 - (1 - (r/c)**2)**3)
    else: 
        return c ** 2 / 6
    
tukey = np.vectorize(tukey)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, tukey(x, 6))
plt.show()
```

```python
true_cps = [50, 75, 20]
found_cps = [54, 75, 88, 104]
found_cps_2 = [53, 78, 74, 71]

def eval(true, found):
    score = 0
    for t in np.array(found):
        errors = np.abs(true- t)
        score += sum(tukey(errors, 6)) - 6*(len(true)-1)
    return score
        
eval(true_cps, found_cps), eval(true_cps, found_cps_2)
```
